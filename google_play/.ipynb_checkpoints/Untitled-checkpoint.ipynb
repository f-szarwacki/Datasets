```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
apps= pd.read_csv('googleplaystore.csv')
reviews = pd.read_csv('googleplaystore_user_reviews.csv')
```

```python
apps.info()
```

```python
reviews.info()
```

```python
apps.sort_values(by = ['Rating'], ascending=False).head()
```

```python
apps.dropna(inplace=True)
```

```python
def normalizeReviews(reviewCount):
    try:
        reviewNum = int(float(reviewCount))
    except:
        reviewNum = "E" + reviewCount
    return reviewNum
```

```python
A = apps['Reviews'].apply(normalizeReviews)
```

```python
apps['Review Count'] = apps['Reviews'].apply(normalizeReviews)
```

```python
def priceNormalize(price):
    if (price[0] == '$'):
        priceNum = float(price[1:])
    elif (price == "0"):
        priceNum = 0
    else:
        priceNum = -1
        print('Error occured')
    return priceNum
```

```python
apps['Price'] = apps['Price'].apply(priceNormalize)
```

```python
apps['Type'].value_counts()
```

```python
apps
```

```python
apps['mega or sth else'] = apps['Size'].apply(lambda x: x[-1])
```

```python
apps['mega or sth else'].value_counts()
```

```python
def installsNormalize(x):
    try:
        installs = int(float(x))
    except:
        if (x[-1] == '+'):
            installs = int(float(x[:-1].replace(',', '')))
        else:
            print('error')
            installs = 0
    return installs
```

```python
apps['Install Count'] = apps['Installs'].apply(installsNormalize)
```

```python
apps['Install Count'].value_counts()
```

```python
apps
```

```python
interest = apps.drop(axis = 1, columns=['Installs', 'Last Updated', 'Current Ver', 'Android Ver', 'mega or sth else'])
```

```python
updates = apps[['Review Count', 'Install Count', 'Rating', 'Last Updated', 'Type']]
```

```python
updates['Log of install count'] = updates['Install Count'].apply(np.log10)
```

```python
updates['Last Update'] = updates['Last Updated'].apply(lambda x: pd.to_datetime(x))
```

```python
updates.sort_values(by = 'Last Update', ascending = False)
```

```python
dateOfReference = pd.Timestamp('2018-08-08')
```

```python
def howManyMonthsHavePassedSinceLastUpdate(date):
    dt = dateOfReference - date
    return int((dt.days / 30))
```

```python
howManyMonthsHavePassedSinceLastUpdate(pd.Timestamp('2018-01-12'))
```

```python
updates['Months from last update'] = updates['Last Update'].apply(howManyMonthsHavePassedSinceLastUpdate)
```

```python
plt.figure(figsize = (20, 10))
ax = sns.boxplot(data = updates, x = 'Install Count', y = 'Months from last update')
ax.set_yticklabels(ax.get_yticklabels(), rotation = 40)
plt.tight_layout()
plt.show()
```
